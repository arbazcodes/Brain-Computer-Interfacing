```python
import mne
import matplotlib.pyplot as plt
import numpy as np
```

```python
edf_file = "varbaz12.edf"
raw = mne.io.read_raw_edf(edf_file, preload=True)
```

```python
raw.ch_names
```

```python
selected_channels = ["F3", "F4"]
raw.pick_channels(selected_channels)
```

```python
start_time = 6.0
end_time = 9.0
raw_crop = raw.copy().crop(tmin=start_time, tmax=end_time)
data = raw_crop.get_data().astype('float64')
times = raw_crop.times
```

```python
plt.figure(figsize=(10, 6))
for i, channel in enumerate(selected_channels):
    plt.plot(times, data[i] , label=channel)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (V)')
plt.title('EEG Data')
plt.legend()
plt.show()
```
